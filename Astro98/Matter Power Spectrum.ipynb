```python
from matplotlib import pyplot as plt
import h5py
import numpy as np
import matplotlib.gridspec as gridspec
import random
from astroML.correlation import two_point

cdm = np.genfromtxt('Data\Ly-a_cdm_sim_model_matterpower.dat', usecols = (0,1))
wdm = np.genfromtxt('Data\ps_ndxf34_wmap7_alpha0.0129703.dat', usecols = (0,1))

print(cdm,wdm)
```

```python
fig = plt.figure(figsize = (10,8))
plt.plot(np.log10(cdm[:,0]), np.log10((cdm[:,1]*(cdm[:,0])**3)/(2*(np.pi)**2)), label = "Cold Dark Matter", linewidth = 5, color = "blue")
plt.plot(np.log10(wdm[:,0]), np.log10((wdm[:,1]*(wdm[:,0])**3)/(2*(np.pi)**2)), label = "Sterile Neutrino, 7 keV", linewidth = 5, color = "red")
plt.ylabel("Log[$\Delta^2$ = k$^3$P(k)/2$\pi^2$]", fontsize = 32)
plt.xlabel("Log(k/(h Mpc$^{-1}$))", fontsize = 32)
plt.legend(fontsize = 16)
plt.tick_params(labelsize=16)
plt.xlim(0, 2.5)
plt.ylim(-4, 2)
plt.savefig("Images/Matter.pdf")
plt.savefig("Images/Matter.png")
plt.show()
```
