```python
#Set up plotting libraries
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib notebook

import numpy as np
import os
import h5py
import matplotlib.gridspec as gridspec
import random
from astroML.correlation import two_point

cdm = h5py.File('Data/COLOR_CDM_DM_subHaloes_z0.00.hdf5', 'r')
wdm = h5py.File('Data/COLOR_WDM_DM_subHaloes_z0.00.hdf5', 'r')

print(cdm,wdm)
```

```python
img_list=[]
file_list = ['Corrfunc_Largest_Sorted', 'Corrfunc_Largest_Sorted2', 'Corrfunc_Largest_Sorted3',
             'Corrfunc_Largest_Sorted_Centrals', 'Corrfunc_Largest_Sorted_Centrals2', 'Corrfunc_Largest_Sorted_Centrals3']
for i in range(0,2):
    img_list.append(mpimg.imread('Images/Data2/' + file_list[i] +'.pdf'))
    
for i in range(3,5):
    img_list.append(mpimg.imread('Images/Data1/' + file_list[i] +'.pdf'))
```

```python
fig, ax = plt.subplots(nrows=2,ncols=3)
ax = ax.flatten()

string_list = ['HII > 0.1', 'HII > 0.1',
               'H2 > 0.1', 'H2 > 0.1', ' ', ' ']

for image, i, string in zip(img_list, range(5), string_list):
        im = ax[i].imshow(image)
        ax[i].axis('off')
        ax[i].text(s = string, x = 10, y = 1150, color = 'white', fontsize = 6)

ax[0].set_title('Face On', fontsize = 10)
ax[1].set_title('Edge On', fontsize = 10)

savefig_args = {}
savefig_args['bbox_inches']='tight'
savefig_args['pad_inches']=0

#fig.set_facecolor('k')
fig.set_size_inches(5.5, 3.5)
fig.savefig('Plot',**savefig_args)

fig.tight_layout(pad=0,w_pad=0,h_pad=0)
fig.subplots_adjust(wspace=0.0,hspace=0.05)
```

```python
CUTE_cdm_sorted = np.genfromtxt("corr_cdm_pos_sorted2.txt", usecols = (0,1))
CUTE_wdm_sorted = np.genfromtxt("corr_wdm_pos_sorted2.txt", usecols = (0,1))

grid = gridspec.GridSpec(2,1, height_ratios = [4,1])
grid.update(hspace = 0)
grid0 = plt.subplot(grid[0])
grid1 = plt.subplot(grid[1])

grid0.loglog(CUTE_cdm_sorted[:,0], CUTE_cdm_sorted[:,1], label = "Cold Dark Matter")
grid0.loglog(CUTE_wdm_sorted[:,0], CUTE_wdm_sorted[:,1], label = "Sterile Neutrino, 7 keV")
grid0.set_xlim(10**-2, 3e1)
grid0.set_ylim(10**-1.25, 10**2)
grid0.set_ylabel(r"$\xi$/Mpc", fontsize = 16)
grid0.legend()
grid0.minorticks_on()
grid0.set_title("0.1 Halo/Mpc$^{3}$", fontsize = 16)

grid1.semilogx(CUTE_wdm_sorted[:,0], CUTE_wdm_sorted[:,1]/CUTE_cdm_sorted[:,1])
grid1.set_xlabel("Radius/Mpc", fontsize = 16)
grid1.set_ylabel("Ratio")
grid1.axhline(y = 1, color = "grey", ls = "--")
grid1.set_xlim(10**-2, 3e1)
grid1.set_ylim(0.8, 1.2)
grid1.minorticks_on()

plt.show()
```
